```octave
function [X_bar, mu] = zeroMean(X)
  mu = mean(X);
  N = size(X)(1);
  X_bar = X - repmat(mu, N, 1);
end
```

```octave
[X_bar, mu] = zeroMean(X)
```

```octave
plot(X_bar(:,1), X_bar(:,2), 'x')
```

```octave
## X is an input matrix of size NxD
## d is number of principal components we wish to keep
## theta is a matrix of size DxD, each column is a principal component
## Z is a matrix of size NxD, the re-representation of X
function [theta, Z] = pca(X, d)
  [X_bar, mu] = zeroMean(X);
  N = size(X, 1);
  S = X_bar' * X_bar / (N-1);
  [theta, lambdas] = eig(S);
  v = diag(lambdas);
  [~, order] = sort(-v);
  theta = theta(:, order(1:d));
  Z = X_bar * theta;
end
```

```octave
[theta, Z] = pca(X, 2)
```
